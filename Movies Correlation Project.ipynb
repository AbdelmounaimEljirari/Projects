```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read in the data

df = pd.read_csv(r'C:\Users\tt\Downloads\movies.csv')
```

# Missing values

```python
# Missing values
count = 0
for col in df.columns:
    if col is None :
      count+=1
print('There are', count, 'Missing values.')
```

# Changing data types

```python
# changing data types 

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df.info()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

# Studying correlation

```python
df.corr()
```

```python
correlation = df.corr()

sns.heatmap(correlation, annot=True)

plt.show()
```

```python
corr_pairs = correlation.unstack()

high_corr = corr_pairs[(corr_pairs > 0.5) & (corr_pairs < 1)]

high_corr
```

# We have Three strong correlation :
Budget - Gross
Gross - Votes
Budget - Votes

```python
plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
sns.regplot(x='budget', y='gross', data = df, line_kws={"color":"black"})
```
